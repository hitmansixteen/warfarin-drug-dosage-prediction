```python
import pandas as pd

# Assuming your dataset is a DataFrame called `df`
# Replace this with your actual DataFrame if it's named differently

Q1 = df['Therapeutic Dose of Warfarin'].quantile(0.25)
Q3 = df['Therapeutic Dose of Warfarin'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_cleaned = df[(df['Therapeutic Dose of Warfarin'] >= lower_bound) & (df['Therapeutic Dose of Warfarin'] <= upper_bound)]

print(f"✅ Filtered rows based on dosage range 20 to 40 mg.")
print(f"📊 Original rows: {len(df)} → Filtered rows: {len(df_cleaned)}")
```

```python
df = df_cleaned
```

```python
# 1. Identify non-numeric columns
categorical_cols = df.select_dtypes(include='object').columns

# 2. One-hot encode them (NaNs will be handled automatically)
df = pd.get_dummies(df, columns=categorical_cols, dummy_na=False)
```

```python
df.isnull().sum()
```

```python
# Replace NaN values with the mean of each column (for numerical columns)
df = df.apply(lambda x: x.fillna(x.mean()) if x.dtype == 'float64' or x.dtype == 'int64' else x, axis=0)
```

```python
# Print unique data types across all columns
print(df.dtypes.unique())
```

```python
X = df.drop(columns=['Therapeutic Dose of Warfarin'])  # Features (input)
y = df['Therapeutic Dose of Warfarin']  # Target (output)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

import pandas as pd
import numpy as np
```

```python
# Now split the PCA-transformed data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42
)
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42),
}
```

```python
# 4. Optionally include XGBoost if available
try:
    from xgboost import XGBRegressor
    models['XGBoost'] = XGBRegressor(objective='reg:squarederror', random_state=42)
except ImportError:
    print("XGBoost not installed. Skipping..."
        )
```

```python
results = []

for name, model in models.items():
    print(f"\n🔄 Training and evaluating: {name}")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"✅ {name} Results:")
    print(f"   - MAE: {mae:.4f}")
    print(f"   - MSE: {mse:.4f}")
    print(f"   - R² Score: {r2:.4f}")

    results.append({
        'Model': name,
        'MAE': mae,
        'MSE': mse,
        'R² Score': r2
    })

# Optionally print the full sorted results table
results_df = pd.DataFrame(results).sort_values(by='R² Score', ascending=False)
print("\n📊 Final Model Comparison:\n")
print(results_df)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, Callback
from sklearn.metrics import r2_score

from sklearn.preprocessing import StandardScaler

# 1. Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Split the scaled data
X_train_dl, X_test_dl, y_train_dl, y_test_dl = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


# 1. Custom Callback to track R² after each epoch
class R2Tracker(Callback):
    def __init__(self, X_val, y_val):
        super().__init__()
        self.X_val = X_val
        self.y_val = y_val
        self.r2_scores = []

    def on_epoch_end(self, epoch, logs=None):
        y_pred = self.model.predict(self.X_val, verbose=0)
        r2 = r2_score(self.y_val, y_pred)
        self.r2_scores.append(r2)
        print(f"📈 Epoch {epoch+1} R² Score: {r2:.4f}")

# 2. Build the model with Batch Normalization
model = Sequential([
    Dense(1024, input_shape=(X_train_dl.shape[1],)),
    BatchNormalization(),
    tf.keras.layers.Activation('relu'),
    Dropout(0.2),

    Dense(512),
    BatchNormalization(),
    tf.keras.layers.Activation('relu'),
    Dropout(0.2),

    Dense(256),
    BatchNormalization(),
    tf.keras.layers.Activation('relu'),
    Dropout(0.2),

    Dense(128),
    BatchNormalization(),
    tf.keras.layers.Activation('relu'),
    Dropout(0.2),

    Dense(64),
    BatchNormalization(),
    tf.keras.layers.Activation('relu'),

    Dense(1)  # Output for regression
])

# 3. Compile
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# 4. Setup early stopping + R² tracking
early_stop = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
r2_tracker = R2Tracker(X_test_dl, y_test_dl)

# 5. Train
history = model.fit(
    X_train_dl, y_train_dl,
    validation_split=0.1,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop, r2_tracker],
    verbose=1
)

# 6. Evaluate model
test_loss, test_mae = model.evaluate(X_test_dl, y_test_dl)
print(f"\n✅ Final Evaluation:")
print(f"   - Test MAE: {test_mae:.4f}")
print(f"   - Test MSE: {test_loss:.4f}")
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['mae'], label='Train MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Training and Validation MAE')
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error')
plt.legend()
plt.grid(True)
plt.show()
```
